```python
import faker
import pandas as pd
import os
```

# 1. 列数据：[学号 班级 姓名 性别 英语 体育 军训 数分 高代  解几]
# 2. 生成1000条数据
# 3. 学号：2308024 + 3位随机数
# 4. 班级：230802 + 2位随机数
# 5. 姓名：faker生成
# 6. 性别：随机生成
# 7. 英语：40-100随机生成
# 8. 体育：40-100随机生成
# 9. 军训：40-100随机生成
# 10. 数分：40-100随机生成
# 11. 高代：40-100随机生成
# 12. 解几：40-100随机生成

```python
# 删除已有的文件
if os.path.exists('stu_score.xlsx '):
    os.remove(' stu_score.xlsx ')
```

```python
f = faker.Faker(locale='zh_CN')
```

```python
# 生成表头
```

```python
columns = ['学号', '班级', '姓名', '性别', '英语', '体育', '军训', '数分', '高代', '解几']
```

# 生成学号

```python
stu_id = []
for i in range(1000):
    stu_id.append('2308024' + str(f.random_int(min=0, max=999, step=1)).zfill(3))
stu_id
```

```python
# 生成班级
```

```python
stu_class = []
for i in range(1000):
    stu_class.append('230802' + str(f.random_int(min=0, max=99, step=1)).zfill(2))
stu_class
```

# 生成姓名

```python
stu_name = []
for i in range(1000):
    stu_name.append(f.name())
```

# 生成性别


```python
stu_sex = ['男' if f.random_int(min=0, max=1, step=1) == 1 else '女' for i in range(1000)]
```

# 生成六科成绩

```python
stu_score = []
for i in range(1000):
    stu_score.append([f.random_int(min=40, max=100, step=1) for i in range(6)])
stu_score
```

# 生成DataFrame

```python
df = pd.DataFrame(columns=columns)
df['学号'] = stu_id
df['班级'] = stu_class
df['姓名'] = stu_name
df['性别'] = stu_sex
df['英语'] = [i[0] for i in stu_score]
df['体育'] = [i[1] for i in stu_score]
df['军训'] = [i[2] for i in stu_score]
df['数分'] = [i[3] for i in stu_score]
df['高代'] = [i[4] for i in stu_score]
df['解几'] = [i[5] for i in stu_score]
```

# 保存到本地

```python
df.to_excel('stu_score.xlsx', index=False)
```

# 读取本地文件

# 随机替换军训和体育列的数据为[缺考 作弊 旷课]

```python
df['解几'] = df['解几'].apply(lambda x: f.random_element(elements=('缺考', '作弊', '旷课')) if f.random_int(min=0, max=300, step=1) == 1 else x)
df['高数'] = df['高数'].apply(lambda x: f.random_element(elements=('缺考', '作弊', '旷课')) if f.random_int(min=0, max=300, step=1) == 1 else x)
df['英语'] = df['英语'].apply(lambda x: f.random_element(elements=('缺考', '作弊', '旷课')) if f.random_int(min=0, max=300, step=1) == 1 else x)
```

# 保存到本地

```python
df.to_excel('stu_score.xlsx', index=False)
```
